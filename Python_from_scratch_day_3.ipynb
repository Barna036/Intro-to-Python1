```python
def word_freq(text):
    words_list = text.split()
    freq = {}
    for word in words_list:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq

string = " I dont care that monday's blue, tuesdays gray"
word_freq(string)
```

Imagine we had a list of employees and salaries from some company, and we wanted to extract a new dictionary of high earners (specifically, those earning more than the mean salary at the company).

Write a function to do this. Then, test it out with a dictionary you create.

```python
salaries = {"Jane": 100, "Mike": 80, "Ali": 85, "Diego": 80}
def employees(salaries):
    mean_salary = sum(salaries.values()) / len(salaries)
    high_earners = {name: salary for name, salary in salaries.items() if salary > mean_salary}
    return high_earners
employees(salaries)

```

```python
salaries = {"Jane": 100, "Mike": 120, "Ali": 85, "Diego": 80}
def high_earners(salaries):
    avg = sum(salaries.values()) / len(salaries)
    high_earners = {}
    for name, salary in salaries.items():
        if salary > avg:
            high_earners[name] = salary
    return high_earners
high_earners(salaries)
```
